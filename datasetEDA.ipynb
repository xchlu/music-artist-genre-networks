```python
import pandas as pd
```

```python
genre_df = pd.read_csv('data/data_by_genres.csv')
artist_df = pd.read_csv('data/data_by_artist.csv')
artist_w_genre = pd.read_csv('data/data_w_genres.csv')
```

```python
categorical_columns = ['mode', 'key']
numerical_columns = ['acousticness', 'danceability', 'duration_ms', 'energy',
                      'instrumentalness', 'liveness', 'loudness', 'speechiness',
                      'tempo','valence','popularity']
```

```python
top10_genres = genre_df.nlargest(10, 'popularity')
top10_genres
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# The musical features of top 10 genres
features = ['acousticness', 'danceability', 'energy', 'liveness']

melted_data = pd.melt(top10_genres, 
                      id_vars=['genres'], 
                      value_vars=features,
                      var_name='feature', 
                      value_name='value')

plt.figure(figsize=(12, 6))
ax = sns.barplot(x='genres', y='value', hue='feature', data=melted_data)

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=3)

ax.set_title('Top 10 Genres by Popularity')
ax.set_ylabel('Values')
ax.set_ylim(0, 1.2)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# Boxplots for numerical features
import numpy as np
fig = plt.figure(figsize=(12, 8))

for i, column in enumerate(numerical_columns):
    ax = plt.subplot(4, 3, i+1)

    box = plt.boxplot(genre_df[column])
    median = np.median(genre_df[column])

    plt.text(1.1, median, f'{median:.2f}', verticalalignment='center')
    plt.title(column)

plt.tight_layout()
plt.show()
```

```python
# Bar charts for categorical features
fig = plt.figure(figsize=(12, 6))

for i, column in enumerate(categorical_columns):
    ax = plt.subplot(1, 2, i+1)

    counts = genre_df[column].value_counts()
    bars = counts.plot(kind='bar', ax=ax)

    for j, v in enumerate(counts):
        ax.text(j, v + 0.1, str(v), ha='center', va='bottom')

    plt.title(column)
    plt.xlabel(column)
    plt.ylabel('Count')

    plt.xticks(rotation=0)

    
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 设置 Seaborn 样式
sns.set(style="whitegrid")

# 查看艺术家数量最多的前10个流派
top_genres = artist_w_genre['genres'].explode().value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette="viridis", hue=None, legend=False)
plt.title('Top 10 Genres by Number of Artists')
plt.xlabel('Number of Artists')
plt.ylabel('Genre')
plt.show()

# 查看艺术家数量最多的前10个艺术家
top_artists = artist_w_genre['artists'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_artists.values, y=top_artists.index, palette="magma", hue=None, legend=False)
plt.title('Top 10 Artists by Number of Tracks')
plt.xlabel('Number of Tracks')
plt.ylabel('Artist')
plt.show()


# 查看音响特征的分布情况
features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence', 'popularity']
artist_w_genre[features].hist(bins=15, figsize=(15, 10), layout=(5, 2), color='skyblue', edgecolor='black')
plt.suptitle('Distribution of Audio Features', size=20)
plt.show()

# 查看不同流派的平均音响特征
mean_features_by_genre = artist_w_genre.explode('genres').groupby('genres')[features].mean().sort_values(by='popularity', ascending=False).head(10)
plt.figure(figsize=(15, 10))
sns.heatmap(mean_features_by_genre, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Mean Audio Features by Top 10 Genres')
plt.show()
```
